Break & Continue

```python
student=["Abubakar", "Ahmad", "Mughal", "Basil", "Taha"]
for i in student:
  if i == "Abubakar":
    break;
  print(i)
```

```python
student=["Abubakar", "Ahmad", "Mughal", "Basil", "Taha"]
for i in student:
  if i == "Abubakar":
    continue;
  print(i)
```
