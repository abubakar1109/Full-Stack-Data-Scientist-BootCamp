Import the Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load a dataset

```python
mydata=pd.read_csv("/content/pokemon.csv")
```

```python
mydata.head()
```

```python
sns.lmplot(x="sp_attack", y='sp_defense', data=mydata)
plt.show()
```

No Regression Line & Adding Hue

```python
sns.lmplot(x="sp_attack", y='sp_defense', data=mydata, fit_reg=False)
plt.show()
```

```python
sns.lmplot(x="sp_attack", y='sp_defense', data=mydata, fit_reg=False, hue='is_legendary')
plt.show()
```

```python
mydata['is_legendary'].value_counts()
```

```python
fig=plt.figure()
a1=fig.add_axes([0,0,1,1])

x=np.arange(1,10)
a1.plot(x,np.exp(x),'b')

a1.set_title('range of No')
plt.ylim(0,1000)
plt.xlim(0,10)

plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=mydata, orient='h')
```

```python
data=mydata.drop(['experience_growth','generation','is_legendary'], axis=1)
sns.boxplot(data=data, orient='h')
```

```python
mydata.columns
```

```python
corr=data.corr(numeric_only=True)
```

```python
corr
```

```python
sns.heatmap(corr)
```

Univariate Visualization

```python
sns.displot(data['defense'], kde=True)
```

```python
sns.boxplot(data['defense'])
```

```python
sns.kdeplot(data['defense'], shade=True)
```

Bivariate Visualization

```python
data.head()
```

```python
sns.jointplot(x='defense', y='attack', data=mydata)
plt.show()
```

```python
mydata[['defense', 'attack']].corr()
```

```python
sns.jointplot(x='defense', y='attack', data=mydata, kind='kde', shade=True)
plt.show()
```

```python
sns.jointplot(x='defense', y='attack', data=mydata, kind='reg')
plt.show()
```

Pairplot

```python
sns.pairplot(data[['defense', 'attack', 'hp']], kind='scatter')
plt.show()
```

```python
sns.pairplot(data[['defense', 'attack', 'hp']], diag_kind='kde')
plt.show()
```
