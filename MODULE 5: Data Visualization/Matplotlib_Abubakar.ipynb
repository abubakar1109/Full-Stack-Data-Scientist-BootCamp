Plot Multiple Point

```python
x=np.array([1,3,5,7,9])
y=x/2

plt.figure(figsize=(5,5))
plt.scatter(x,y,c='green')
plt.show()
```

Label the Axes

```python
plt.plot([1,2,3,4])
plt.ylabel('Y-Label')
plt.xlabel('X-Label')
plt.show()
```

Create Multiple plot with subplot

```python
friend=['abubakar','ahmad','mughal','basil','taha']
rno=[1,7,2,9,3]

plt.figure(figsize=(15,5))

plt.subplot(131)
plt.bar(friend, rno)

plt.subplot(132)
plt.scatter(friend, rno)

plt.subplot(133)
plt.plot(friend, rno)

plt.suptitle('Categorical Plotting')
plt.show()
```
