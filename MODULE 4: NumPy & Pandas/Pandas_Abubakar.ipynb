Join & Merge

```python
friends={'abubakar':10000,
         'ahamd':5000,
         'basil':440,
         'mughal':6700,
         'taha':300}
area={'abubakar':'hassanabdal',
         'ahamd':'overseas',
         'arayash':np.nan,
         'basil':'bahria',
         'ar':np.nan,
         'mughal':'chor',
         'taha':'wah'}
```

```python
frienddf=pd.DataFrame.from_dict(friends, orient='index',columns=['friend'])
areadf=pd.DataFrame.from_dict(area, orient='index',columns=['area'])
```

```python
frienddf
```

```python
areadf
```

```python
joined_df=areadf.join(frienddf, how='left')
```

```python
joined_df
```

```python
joined_df=areadf.join(frienddf, how='right')
```

```python
joined_df
```

```python
joined_df=areadf.join(frienddf, how='inner')
```

```python
joined_df
```

```python
joined_df=areadf.join(frienddf, how='outer')
```

```python
joined_df
```

Merge

```python
frienddf.index.name='names'
areadf.index.name='names'
```
