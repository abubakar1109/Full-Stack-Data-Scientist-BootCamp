Manipulating Data Using Pandas

```python
print('Data Types:')
data.info()
```

```python
print('Descriptive Statistics for Numerical Columns:')
display(data.describe().T)

print('\nValue Counts for Categorical Columns:')
for column in ['sex', 'smoker', 'region']:
    print(f'\n{column}:')
    display(data[column].value_counts())
```

```python
print('Average Expenses by Region and Smoker Status:')
display(data.groupby(['region', 'smoker'])['expenses'].mean().unstack())
```

```python
filtered_data = data[(data['sex'] == 'female') & (data['smoker'] == 'no') & (data['region'] == 'northwest')]
print('Non-smoking females in the northwest region (first 5 rows):')
display(filtered_data.head())
```

```python
bins = [18, 30, 45, 60, 100]
labels = ['18-29', '30-44', '45-59', '60+']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

print('DataFrame with new age_group column (first 5 rows):')
display(data[['age', 'age_group', 'expenses']].head())
```

```python
pivot_table_expenses = pd.pivot_table(data, values='expenses',
                                    index=['region'],
                                    columns=['sex', 'smoker'],
                                    aggfunc=np.mean)

print('Average expenses by region, sex, and smoker status:')
display(pivot_table_expenses)
```
