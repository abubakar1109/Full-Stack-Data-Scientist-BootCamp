```python
sales = {'Ahmad': 10000,
         'Mughal': 5000,
         'Basil': 440,
         'Taha': 6700,
         'Quershi': 300
        }

region = {'Ahmad': 'West',
          'Mughal': np.nan,
          'Taha': 'West',
          'Abdur Rehman': np.nan,
          'Ayesha': 'East',
          'Araysh': 'South',
          'Saira': 'West',
        }
```

```python
sales_df = pd.DataFrame.from_dict(sales, orient='index',
                                  columns=['sales'])
region_df = pd.DataFrame.from_dict(region, orient='index',
                                   columns=['region'])
```

```python
sales_df
```

```python
region_df
```

```python
Left_join_df = region_df.join(sales_df, how='left')
Right_join_df = region_df.join(sales_df, how='right')
inner_join_df = region_df.join(sales_df, how='inner')
outer_join_df = region_df.join(sales_df, how='outer')
```

```python
print('Left Join\n',Left_join_df)
print('Right Join\n',Right_join_df)
print('Inner Join\n',inner_join_df)
print('Outer Join\n',outer_join_df)
```

```python
region_df.index.name='names'
sales_df.index.name='names'
```

```python
print(pd.merge(region_df,sales_df,on='names',how='right'))
```
